# First measurement
- describe here the first sample

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

lw = 1.3
ms = 3

# Enter the time passed when measurements are taken since the first one in [s]
n = 10      # number of measurement taken
t = c = np.linspace(120,120*n,n)
t = np.array(t,dtype=float)

# enter the value of the mass measured every Dt in [g]
Mcont = 123.120;     # cointainer mass in [g]
Mw = [15.011,14.980,14.953,14.928,14.896,14.862,14.828,14.801,14.775,14.742]
Mw = np.array(Mw,dtype=float)

E = np.zeros(n)
# Load image
for k in range(0,n):
    E[k] = (Mw[k] - Mw[k-1])/(t[k] - t[k-1])
    print(f"Evaporation flux, E = {E[k]:.4g}")
    print(f"index, k ={k}")
    
df = pd.DataFrame({"time, t [s]": t, "Evaporation flux, E [g / s]": E})
df

# Plot
plt.figure(1)
plt.plot(t,Mw,'r.',linewidth=lw,markersize=ms)
plt.xlabel(r"time, $t$ [s]")
plt.ylabel(r"Measured mass, $M$ [g/s]")
plt.grid(False)
plt.tight_layout()
plt.show()


# Plot
plt.figure(2)
plt.plot(t,E,'r.',linewidth=lw,markersize=ms)
plt.xlabel(r"time, $t$ [s]")
plt.ylabel(r"Evaporation flux, $E$ [g/s]")
plt.grid(False)
plt.tight_layout()
plt.show()

Eav = np.mean(E)
print(f"Averaged evaporation flux ={Eav:.4g} g/s")
```
